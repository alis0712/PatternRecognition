# Problem 1 a), c)

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal
from numpy.linalg import det, inv
from numpy.random import seed, randint, random
from sklearn import datasets
```

# Our 2 dimensional will be over variables x and y

```python
N = 1000
X = np.linspace(-4, 4, N)
Y = np.linspace(-4, 4, N)
X, Y = np.meshgrid(X, Y)
```

#  Mean Vector and Covariance Matrix

```python
mu = np.array([1 , 2])
sigma = np.array([[4, 4], [4 , 9]])
```

# Pack X and Y into a single 3-dimensional array

```python
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

def multivariate_gaussian(pos, mu , sigma):
    """Return the multivariate Gaussian distribution on array pos. pos is an array constructed by packing the meshed arrays of variables 
    x_1, x_2,x_3,....,x_k into its _last_dimension."""
    
    n = mu.shape[0]
    sigma_det = np.linalg.det(sigma)
    sigma_inv = np.linalg.inv(sigma)
    N = np.sqrt((2*np.pi)**n * sigma_det)
# This einsum call calculates (x-mu)T.sigma-1.(x-mu) in a vectorized way across all the input variables.
    fac = np.einsum('...k,kl,...l->...', pos-mu, sigma_inv,pos-mu)
    return np.exp(-fac / 2) / N
# The distribution on the variables X,Y packed into pos.
Z = multivariate_gaussian(pos, mu, sigma)
print(Z.shape) 
```

# Create a surface plot and projected filled contour plot under it

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z, rstride = 3, 
                cstride = 3, linewidth = 1, 
                antialiased = True, cmap = cm.viridis)
```


# Adjust the limits, ticks, and view angle

```python
ax.set_zlim(-0.15, 0.2)
ax.set_zticks(np.linspace(0, 0.2, 5))
ax.view_init(27, -21)

plt.show()
```

# Problem 1 d) Given Parameters

```python
mu = np.array([5 , -5,  6])
sigma = np.array([[5, 2, -1], 
                  [2 , 5, 0], 
                  [-1, 0, 4]])
```

# Let us evaluate the integral using numerical and monte-carlo method

```python
from scipy.stats.mvn import mvnun

cum = mvnun(np.array([-np.inf, -np.inf, -np.inf]), np.array([np.inf, np.inf, np.inf]), mu, sigma)
print(cum[0])

def multpdf(X, mu, sigma, invsig):
    d = X.shape[0]
    den = (2*np.pi)**(d/2.0)*np.sqrt(det(sigma))
    p = np.exp(-0.5*np.dot((X-mu).T, np.dot(invsig, (X-mu))))/den
    return(p)

count = 0
nrandV = 100000
invsig = sigma
for i in range(nrandV):
    x1 = randint(-10, 10)
    x2 = randint(-10, 10)
    x3 = randint(-10, 10)
    p1 = random() * 0.0635
    X = np.array([x1, x2, x3])
    p = multpdf(X.T, mu.T, sigma, invsig)
    if(p1 < p):
        count += 1
        
vol = (count/nrandV) * 0.0635 * 20 * 20 * 20
print(vol)
```

# Problem 2 Iris data classification using Baysian decision 

# Import some data to play with

```python
iris = datasets.load_iris()
X = iris.data 
y = iris.target
print(X.shape)
print(y.shape)

```

# Compute the mean value mue = E(X) for all three classes 0, 1, 2 as 2D-array, each column has mean vector class

```python
mu = np.zeros(shape=(3,4), dtype = 'float')
mu1 = np.zeros(4, dtype ='float')
mu2 = np.zeros(4, dtype ='float')
mu3 = np.zeros(4, dtype ='float')
XX1 = X[0:25,]
XX2 = X[25:50,]
XX3 = X[50:75,]
for i in range(4):
    mu1[i] = np.mean(XX1[:,i])
    mu2[i] = np.mean(XX2[:,i])
    mu3[i] = np.mean(XX3[:,i])
mu[0,] = mu1
mu[1,] = mu2
mu[2,] = mu3

print(mu)


    
```

```python
Bsig = np.zeros(shape=(3,4,4), dtype = 'float')
Bsig1 = np.cov(XX1.T)
Bsig2 = np.cov(XX2.T)
Bsig3 = np.cov(XX3.T)
Bsig[0,] = Bsig1
Bsig[1,] = Bsig2
Bsig[2,] = Bsig3
print(Bsig.shape)
print(Bsig[0])
print('\n', Bsig[1])
print('\n', Bsig[2])
iBsig = np.zeros(shape=(3,4,4), dtype = 'float')
for i in range(3):
    iBsig[i] = np.linalg.inv(Bsig[i])
print(iBsig[0])
print('\n', iBsig[1])
print('\n', iBsig[2])
```

```python
dBsig = np.zeros(3, dtype = 'float')
rdBsig = np.zeros(3, dtype = 'float')
for i in range (3):
    dBsig[i] = np.linalg.det(Bsig[i])
    rdBsig[i] = np.sqrt(dBsig[i])
print(dBsig)
print(rdBsig)
```

# compute probability density of a given vector for p(x,k), where x is input vector and k is class number 0:2

```python
def pdensity(x, m, s, rd):
    den = (2*np.pi)**(x.shape[0]/2) * rd
    return(np.exp(-0.5*np.dot((x-m), np.dot(s, x-m)))/den)
```

# apriors for all three classes are equal = 1/3 and denominirs are equal for all the three classes the Posterior density for each class multiplied by a constant term in the decision can be from probability density

# Let us apply for each vector and find max aposterior value for each vector from the sample

```python
px = np.zeros(3,dtype = 'float')
mcl = 0
for i in range(25):
    x = X[i,]
    for j in range(3):
        px[j] = pdensity(x, mu[j,], Bsig[j,], rdBsig[j]) #p(x) to be divided by sum (p(x/w)*P(w)) for all w
    mx = np.argmax(px)
    if(mx != 0):
        mcl += 1
for i in range (25,50):
    x = X[i,]
    for j in range(3):
        px[j] = pdensity(x, mu[j,], Bsig[j,], rdBsig[j]) #p(x) to be divided by sum (p(x/w)*P(w)) for all w
    mx = np.argmax(px)
    if(mx != 0):
        mcl += 1
for i in range(50, 75):
    x = X[i,]
    for j in range(3):
        px[j] = pdensity(x, mu[j,], Bsig[j,], rdBsig[j]) #p(x) to be divided by sum (p(x/w)*P(w)) for all w
    mx = np.argmax(px)
    if(mx != 0):
        mcl += 1
print('Misclassification = ', mcl, 'Accuracy = ', (75-mcl)*100.0/75)
```

# Split Dataset into (drivetrain, trainlabel,datatest,testlabel) 30X3 samples for training set and 20 X 3 for testing

```python
def splitdata(X,y):
    Xtr = np.zeros(shape=(90,4), dtype = 'float')
    ytr = np.zeros(90, dtype = 'uint16')
    Xts = np.zeros(shape=(60,4), dtype = 'float')
    yts = np.zeros(60, dtype = 'uint16')
    
    Xtr[0:30,] = X[0:30,]
    Xtr[30:60,] = X[50:80,]
    Xtr[60:90,] = X[100:130,]
    
    Xts[0:20,] = X[30:50,]
    Xts[20:40,] = X[80:100,]
    Xts[40:60,] = X[130:150,]
    
    ytr[0:30] = y[0:30]
    ytr[30:60] = y[50:80]
    ytr[60:90] = y[100:130]
    
    yts[0:20,] = y[30:50]
    yts[20:40,] = y[80:100]
    yts[40:60,] = y[130:150]
    
    
    return Xtr, ytr, Xts, yts
```

# Compute the mean value mue = E(X) for all three classes 0, 1, 2 as 2D-array, each column has mean vector class

```python
mu = np.zeros(shape=(3,4), dtype = 'float')
mu1 = np.zeros(4, dtype ='float')
mu2 = np.zeros(4, dtype ='float')
mu3 = np.zeros(4, dtype ='float')

Xtr, ytr, Xts, yts = splitdata(X,y)


for i in range(4):
    mu1[i] = np.mean(Xtr[0:30,i])
    mu2[i] = np.mean(Xtr[30:60,i])
    mu3[i] = np.mean(Xtr[60:90,i])
mu[0,] = mu1
mu[1,] = mu2
mu[2,] = mu3

print(mu)
```
